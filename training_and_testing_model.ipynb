<h1>Training and Testing model using sklearn.model_selection.train_test_split</h1>


```python
import pandas as pd
df = pd.read_csv('carprices.csv')
bmw = df.head()
```

```python
bmw
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
plt.scatter(bmw['Mileage'],bmw['Sell Price($)'])
```

```python
plt.scatter(bmw['Age(yrs)'],bmw['Sell Price($)'])
```

```python
x = bmw[['Mileage','Age(yrs)']]
x
```

```python
y = bmw['Sell Price($)']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

```

```python
len(X_train)
```

```python
from sklearn.linear_model import LinearRegression
bmw_test = LinearRegression()
```

```python
bmw_test.fit(X_train,y_train)
```

```python
bmw_test.predict(X_test)
```

```python
y_test
```

```python
bmw_test.score(X_train,y_train)
```

```python
bmw_test.score(X_test,y_test)
```
